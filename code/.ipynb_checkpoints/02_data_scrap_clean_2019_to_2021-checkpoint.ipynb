```python
# call some libraries
import os
import datetime
import pandas as pd

#Import YouTubeDataAPI
from youtube_api import YouTubeDataAPI
from youtube_api.youtube_api_utils import *
from dotenv import load_dotenv

api_key = " "
yt = YouTubeDataAPI(api_key)
```

```python
video_2019 = pd.DataFrame(yt.search(q='EV, policy', regionCode="US", publishedAfter='2019-01-01T00:00:00Z', publishedBefore='2019-12-31T23:59:59Z',max_results = 2000))
video_2020 = pd.DataFrame(yt.search(q='EV, policy', regionCode="US", publishedAfter='2020-01-01T00:00:00Z', publishedBefore='2020-12-31T23:59:59Z',max_results = 2000))
```

```python
video_2019 = video_2019[0:300].copy()
```

```python
video_2019.to_csv('video_2019.csv',index = False)
```

```python
video_2020 = video_2020[0:301].copy()
```

```python
video_2020.to_csv('video_2020.csv', index = False)
```

```python
videoId_2019 = video_2019['video_id'].tolist()

videoId_2020 = video_2020['video_id'].tolist()
```

```python
list_comments_2019 = []
for video_id in videoId_2019:
    try:
        
        comments_2019 = yt.get_video_comments(video_id,max_results = 500)
        list_comments_2019.append(pd.DataFrame(comments_2019))
    except Exception as e:
        print(f"Error with video ID {video_id}: {e}")
        
# Concatenate all comment DataFrames
comments_2019 = pd.concat(list_comments_2019)
comments_2019
```

```python
comments_2019 = comments_2019.dropna(subset=['video_id'])
comments_2019 = comments_2019[comments_2019['text'].str.len() >= 100]
comments_2019.copy()
```

```python
comments_2019.to_csv('comments_2019_clean.csv',index = False)
```

```python
list_comments_2020 = []
for video_id in videoId_2020:
    try:
        
        comments_2020 = yt.get_video_comments(video_id,max_results = 500)
        list_comments_2020.append(pd.DataFrame(comments_2020))
    except Exception as e:
        print(f"Error with video ID {video_id}: {e}")
        
# Concatenate all comment DataFrames
comments_2020 = pd.concat(list_comments_2020)
comments_2020
```

```python
comments_2020 = comments_2020.dropna(subset=['video_id'])
comments_2020 = comments_2020[comments_2020['text'].str.len() >= 100]
comments_2020.copy()
```

```python
comments_2020.to_csv('comments_2020_clean.csv',index = False)
```
